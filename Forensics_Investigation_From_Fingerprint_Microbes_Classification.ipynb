```python
from sklearn.preprocessing import LabelEncoder
labelencoder_Y = LabelEncoder()
df.iloc[:,1] = labelencoder_Y.fit_transform(df.iloc[:,1].values)

df.iloc[:,1]
```

```python
X = df.drop("hand", axis = 1).values #Features - Independent
Y = df["hand"].values #Target - Dependent

type(X)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler #Importing another library.

sc = StandardScaler() 
X_train = sc.fit_transform(X_train) 
X_test = sc.fit_transform(X_test)
```

```python
X_train
```

```python
from sklearn.neighbors import KNeighborsClassifier
neighbor = KNeighborsClassifier(n_neighbors = 3)
model = neighbor.fit(X_train, Y_train)
```

```python
print(neighbor.score(X_train, Y_train))
```

```python
from sklearn.metrics import confusion_matrix

prediction = model.predict(X_test)

cm = confusion_matrix(Y_test, prediction)
print(cm)
```

```python
#True Positive 
TP = cm[0][0]
#True Negative
TN = cm[1][1]
#False Negative
FN = cm[1][0]
#False Positive
FP = cm[0][1]

print('Test Accuracy: ',(TP + TN) / (TP + TN + FN + FP))
```

```python
#Other metrics of the model:
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score

print(classification_report(Y_test, prediction))
print(accuracy_score(Y_test, prediction))
```
